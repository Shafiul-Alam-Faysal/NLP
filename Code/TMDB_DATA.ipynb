```python
import requests
import pandas as pd
import time

API_KEY = "47e12dae35ebc66fb36b12f1318cd2e6"

# Get genre mapping
genre_url = (
    f"https://api.themoviedb.org/3/genre/movie/list"
    f"?api_key={API_KEY}&language=en-US"
)

genre_data = requests.get(genre_url).json()

genre_map = {
    genre["id"]: genre["name"]
    for genre in genre_data["genres"]
}

movies = []

# Get total pages automatically
first_page = requests.get(
    f"https://api.themoviedb.org/3/movie/top_rated"
    f"?api_key={API_KEY}&language=en-US&page=1"
).json()

total_pages = first_page["total_pages"]

print(f"Total pages: {total_pages}")

for page in range(1, min(total_pages, 500) + 1):

    url = (
        f"https://api.themoviedb.org/3/movie/top_rated"
        f"?api_key={API_KEY}&language=en-US&page={page}"
    )

    data = requests.get(url).json()
    
    if "results" not in data:
        print(f"Problem on page {page}")
        print(data)
        break

    for movie in data["results"]:

        genres = [
            genre_map[g]
            for g in movie["genre_ids"]
            if g in genre_map
        ]

        movies.append({
            "title": movie["title"],
            "overview": movie["overview"],
            "genres": ", ".join(genres)
        })

    print(f"Page {page}/{total_pages}")

    time.sleep(0.1)

df = pd.DataFrame(movies)

df.to_csv("TMDB_DATA.csv", index=False, encoding="utf-8")

print(df.shape)
print("Saved TMDB_DATA.csv")
```
