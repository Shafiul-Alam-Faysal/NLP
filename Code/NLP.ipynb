```python
import os
import pandas as pd
import re
from bs4 import BeautifulSoup
import string, time
```

```python
string.punctuation
```

```python
def clean_text(text):
    text = BeautifulSoup(text, "html.parser").get_text()
    text = re.sub(r"https?://\S+|www\.\S+", "", text)
    text = text.translate(
        str.maketrans("", "", string.punctuation)
    )
    text = re.sub(r"\s+", " ", text)
    text = text.lower()
    return text
```
