```python
import pandas as pd

def get_pos_tags(text):
    doc = nlp(text)

    data = []

    for token in doc:
        data.append({
            "Word": token.text,
            "Tag": token.tag_,
            "Meaning": spacy.explain(token.tag_)
        })

    return pd.DataFrame(data)
```

```python
get_pos_tags(doc)
```

```python
get_pos_tags("I left the room")
```

```python
get_pos_tags("to the left of the room")
```

```python
get_pos_tags("I read books on history")
```

```python
get_pos_tags("I have a read a book history")
```
